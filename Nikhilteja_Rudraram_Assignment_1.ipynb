# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math
def hemisphereVolume(radius):
    return (2/3) * math.pi * (radius ** 3)

print("r\tV")
for radius in range(2, 21, 2):
    volume = hemisphereVolume(radius)
    print(f"{radius}\t{volume:.2f}")


```

```python
# Your code here
import math

def inputExpressionEvaluate(expression):
    try:
        result = eval(expression, {"__builtins__": None}, {"sqrt": math.sqrt,"pow": math.pow, "mod": lambda x, y: x % y})
        return result
    except ZeroDivisionError:
        return "Division by zero is not allowed."
    except (SyntaxError, NameError):
        return "Invalid mathematical expression."
    except Exception as e:
        return f"{str(e)}"

def main():
    print("Type 'exit' to quit the calculator")

    while True:
        expression = input("Enter expression to calculate: ")

        if expression.lower() == 'exit':
            print("Exiting calculator.........")
            break

        result = inputExpressionEvaluate(expression)
        print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
# Your code here
import csv
import os
import logging
from collections import defaultdict

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# File to store inventory data
INVENTORY_FILE = "inventory_data.csv"

class InventorySystem:
    def __init__(self):
        self.inventory = []
        self.load_inventory()

    def load_inventory(self):
        """Load inventory from a CSV file, handling potential corruption."""
        if not os.path.exists(INVENTORY_FILE):
            logging.warning("Inventory file not found. Initializing new inventory.")
            self.initialize_inventory()
            return

        try:
            with open(INVENTORY_FILE, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                self.inventory = [row for row in reader]
        except Exception as e:
            logging.error(f"Error loading inventory: {e}. Initializing new inventory.")
            self.initialize_inventory()

    def initialize_inventory(self):
        """Initialize inventory with 50 sample products."""
        self.inventory = [
            {"ID": str(i+1), "Name": f"Item {i+1}", "Category": f"Category {i%5 + 1}", "Stock": str(10 + i), "Price": str(5 * (i % 10 + 1))}
            for i in range(50)
        ]
        self.save_inventory()

    def save_inventory(self):
        """Save inventory data to a CSV file."""
        try:
            with open(INVENTORY_FILE, mode='w', newline='') as file:
                fieldnames = ["ID", "Name", "Category", "Stock", "Price"]
                writer = csv.DictWriter(file, fieldnames=fieldnames)
                writer.writeheader()
                writer.writerows(self.inventory)
            logging.info("Inventory successfully saved.")
        except Exception as e:
            logging.error(f"Error saving inventory: {e}")

    def add_product(self, name, category, stock, price):
        """Add a new product to the inventory."""
        new_id = str(len(self.inventory) + 1)
        self.inventory.append({"ID": new_id, "Name": name, "Category": category, "Stock": str(stock), "Price": str(price)})
        self.save_inventory()
        logging.info(f"Product added: {name}, Category: {category}, Stock: {stock}, Price: {price}")

    def remove_product(self, product_id):
        """Remove a product by ID."""
        self.inventory = [p for p in self.inventory if p["ID"] != product_id]
        self.save_inventory()
        logging.info(f"Product with ID {product_id} removed.")

    def modify_product(self, product_id, name=None, category=None, stock=None, price=None):
        """Modify product details."""
        for product in self.inventory:
            if product["ID"] == product_id:
                if name:
                    product["Name"] = name
                if category:
                    product["Category"] = category
                if stock:
                    product["Stock"] = str(stock)
                if price:
                    product["Price"] = str(price)
                self.save_inventory()
                logging.info(f"Product {product_id} updated.")
                return
        logging.warning("Product not found.")

    def find_product(self, keyword):
        """Find products by keyword."""
        return [p for p in self.inventory if keyword.lower() in p["Name"].lower() or keyword.lower() in p["Category"].lower()]

    def low_stock_report(self):
        """Generate a report for low stock items."""
        return [p for p in self.inventory if int(p["Stock"]) < 5]

    def category_summary(self):
        """Generate a category breakdown of inventory."""
        category_summary = defaultdict(lambda: {"count": 0, "total_value": 0})
        for product in self.inventory:
            category = product["Category"]
            category_summary[category]["count"] += int(product["Stock"])
            category_summary[category]["total_value"] += int(product["Stock"]) * float(product["Price"])
        return category_summary

# Example usage
if __name__ == "__main__":
    system = InventorySystem()
    system.add_product("New Item", "Category 1", 20, 30)
    print("Search results:", system.find_product("Item 10"))
    print("Low Stock Report:", system.low_stock_report())
    print("Category Breakdown:", system.category_summary())

```

```python
import csv
import os
import logging
from collections import defaultdict

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# File to store inventory data
INVENTORY_FILE = "inventory_data.csv"

class InventorySystem:
    def __init__(self):
        self.inventory = []
        self.load_inventory()

    def load_inventory(self):
        """Load inventory from a CSV file, handling potential corruption."""
        if not os.path.exists(INVENTORY_FILE):
            logging.warning("Inventory file not found. Initializing new inventory.")
            self.initialize_inventory()
            return

        try:
            with open(INVENTORY_FILE, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                self.inventory = [row for row in reader if all(field in row for field in ["ID", "Name", "Category", "Stock", "Price"])]
        except Exception as e:
            logging.error(f"Error loading inventory: {e}. Initializing new inventory.")
            self.initialize_inventory()

    def initialize_inventory(self):
        """Initialize inventory with 50 sample products."""
        self.inventory = [
            {"ID": str(i+1), "Name": f"Item {i+1}", "Category": f"Category {i%5 + 1}", "Stock": str(10 + i), "Price": str(5 * (i % 10 + 1))}
            for i in range(50)
        ]
        self.save_inventory()

    def save_inventory(self):
        """Save inventory data to a CSV file."""
        try:
            with open(INVENTORY_FILE, mode='w', newline='') as file:
                fieldnames = ["ID", "Name", "Category", "Stock", "Price"]
                writer = csv.DictWriter(file, fieldnames=fieldnames)
                writer.writeheader()
                writer.writerows(self.inventory)
            logging.info("Inventory successfully saved.")
        except Exception as e:
            logging.error(f"Error saving inventory: {e}")

    def add_product(self, name, category, stock, price):
        """Add a new product or update stock if it already exists."""
        for product in self.inventory:
            if product["Name"] == name and product["Category"] == category:
                product["Stock"] = str(int(product["Stock"]) + stock)
                logging.info(f"Existing product updated: {name}, New Stock: {product['Stock']}")
                self.save_inventory()
                return

        if not isinstance(stock, int) or not isinstance(price, (int, float)) or stock < 0 or price < 0:
            logging.error("Invalid stock or price values. Ensure they are non-negative numbers.")
            return

        new_id = str(len(self.inventory) + 1)
        self.inventory.append({"ID": new_id, "Name": name, "Category": category, "Stock": str(stock), "Price": str(price)})
        self.save_inventory()
        logging.info(f"Product added: {name}, Category: {category}, Stock: {stock}, Price: {price}")

    def find_product(self, name):
        """Search for a product by name and return matching entries."""
        return [product for product in self.inventory if name.lower() in product["Name"].lower()]

    def low_stock_report(self, threshold=5):
        """Return a list of products with stock below the threshold."""
        return [product for product in self.inventory if int(product["Stock"]) < threshold]

    def category_summary(self):
        """Generate a category breakdown of inventory, ensuring empty categories are included."""
        category_summary = defaultdict(lambda: {"count": 0, "total_value": 0.0})

        for product in self.inventory:
            category = product["Category"]
            category_summary[category]["count"] += int(product["Stock"])
            category_summary[category]["total_value"] += int(product["Stock"]) * float(product["Price"])

        return category_summary

    def remove_product(self, product_id):
        """Remove a product from inventory by ID, handling non-existent IDs."""
        self.inventory = [product for product in self.inventory if product["ID"] != product_id]
        self.save_inventory()
        logging.info(f"Product with ID {product_id} removed.")

    def update_price(self, product_id, new_price):
        """Update the price of a product, ensuring it's a valid number."""
        for product in self.inventory:
            if product["ID"] == product_id:
                if new_price < 0:
                    logging.error("Price cannot be negative.")
                    return
                product["Price"] = str(new_price)
                self.save_inventory()
                logging.info(f"Price updated for product ID {product_id} to {new_price}.")
                return
        logging.error("Product ID not found.")

# Example usage
if __name__ == "__main__":
    system = InventorySystem()
    system.add_product("New Item", "Category 1", 20, 30)
    print("Search results:", system.find_product("Item 10"))
    print("Low Stock Report:", system.low_stock_report())
    print("Category Breakdown:", system.category_summary())
```

```python
# Your code here
import csv
import os

FLIGHT_FILE = "flights.csv"

# Edge cases handled:
# 1. Invalid input format (non-integer for seat count, invalid date format)
# 2. Searching for a flight that doesn't exist
# 3. Booking a seat when the flight is full
# 4. Booking a seat on a non-existent flight
# 5. Releasing a seat that wasn't booked
# 6. Updating a flight that doesn't exist
# 7. Cancelling a flight that doesn't exist
# 8. Duplicate flight entries
# 9. Incorrect file handling errors
# 10. Empty CSV file handling

def initialize_csv():
    if not os.path.exists(FLIGHT_FILE) or os.stat(FLIGHT_FILE).st_size == 0:
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "Date", "TotalSeats", "BookedSeats"])
            for i in range(1, 51):
                writer.writerow([f"FL{i:03}", "CityA", "CityB", "2025-12-10", 100, 0])

def read_flights():
    with open(FLIGHT_FILE, mode='r') as file:
        reader = csv.DictReader(file)
        return list(reader)

def write_flights(flights):
    with open(FLIGHT_FILE, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=["FlightID", "Origin", "Destination", "Date", "TotalSeats", "BookedSeats"])
        writer.writeheader()
        writer.writerows(flights)

def search_flight(flight_id):
    flights = read_flights()
    for flight in flights:
        if flight["FlightID"] == flight_id:
            return flight
    return None

def book_seat(flight_id):
    flights = read_flights()
    for flight in flights:
        if flight["FlightID"] == flight_id:
            if int(flight["BookedSeats"]) < int(flight["TotalSeats"]):
                flight["BookedSeats"] = str(int(flight["BookedSeats"]) + 1)
                write_flights(flights)
                print("Booking successful!")
            else:
                print("Flight is fully booked.")
            return
    print("Flight not found.")

def release_seat(flight_id):
    flights = read_flights()
    for flight in flights:
        if flight["FlightID"] == flight_id:
            if int(flight["BookedSeats"]) > 0:
                flight["BookedSeats"] = str(int(flight["BookedSeats"]) - 1)
                write_flights(flights)
                print("Seat released successfully!")
            else:
                print("No booked seats to release.")
            return
    print("Flight not found.")

def update_flight(flight_id, new_origin, new_destination, new_date):
    flights = read_flights()
    for flight in flights:
        if flight["FlightID"] == flight_id:
            flight["Origin"] = new_origin
            flight["Destination"] = new_destination
            flight["Date"] = new_date
            write_flights(flights)
            print("Flight updated successfully!")
            return
    print("Flight not found.")

def cancel_flight(flight_id):
    flights = read_flights()
    flights = [flight for flight in flights if flight["FlightID"] != flight_id]
    write_flights(flights)
    print("Flight cancelled successfully.")

def performance_report():
    flights = read_flights()
    total_flights = len(flights)
    total_booked = sum(int(flight["BookedSeats"]) for flight in flights)
    print(f"Total Flights: {total_flights}, Total Booked Seats: {total_booked}")

def main():
    initialize_csv()
    while True:
        print("\nFlight Booking System:")
        print("1. Search Flight")
        print("2. Book Seat")
        print("3. Release Seat")
        print("4. Update Flight")
        print("5. Cancel Flight")
        print("6. Performance Report")
        print("7. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            fid = input("Enter Flight ID: ")
            flight = search_flight(fid)
            print(flight if flight else "Flight not found.")
        elif choice == "2":
            fid = input("Enter Flight ID to book seat: ")
            book_seat(fid)
        elif choice == "3":
            fid = input("Enter Flight ID to release seat: ")
            release_seat(fid)
        elif choice == "4":
            fid = input("Enter Flight ID to update: ")
            origin = input("New Origin: ")
            dest = input("New Destination: ")
            date = input("New Date (YYYY-MM-DD): ")
            update_flight(fid, origin, dest, date)
        elif choice == "5":
            fid = input("Enter Flight ID to cancel: ")
            cancel_flight(fid)
        elif choice == "6":
            performance_report()
        elif choice == "7":
            print("Exiting...")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```

```python
import csv
import os
import random
import smtplib
from datetime import datetime

FLIGHT_FILE = "flights.csv"
USERS_FILE = "users.csv"

# Edge cases handled:
# 1. Invalid input format (non-integer for seat count, invalid date format)
# 2. Searching for a flight that doesn't exist
# 3. Booking a seat when the flight is full
# 4. Booking a seat on a non-existent flight
# 5. Releasing a seat that wasn't booked
# 6. Updating a flight that doesn't exist
# 7. Cancelling a flight that doesn't exist
# 8. Duplicate flight entries
# 9. Incorrect file handling errors
# 10. Empty CSV file handling
# 11. User authentication for booking
# 12. Pricing information for flights
# 13. Payment integration (mocked for now)
# 14. Seat selection option
# 15. Confirmation email (mocked for now)
# 16. Round trip booking option
# 17. Search filters (price, duration, stops)
# 18. Customer support option
# 19. Baggage information display
# 20. Refund or rescheduling policy

def initialize_csv():
    if not os.path.exists(FLIGHT_FILE) or os.stat(FLIGHT_FILE).st_size == 0:
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "Date", "TotalSeats", "BookedSeats", "Price", "Stops", "Duration", "BaggageAllowance"])
            for i in range(1, 51):
                writer.writerow([f"FL{i:03}", "CityA", "CityB", "2025-12-10", 100, 0, random.randint(100, 500), random.randint(0, 2), random.randint(1, 10), "20kg"])

def initialize_users():
    if not os.path.exists(USERS_FILE) or os.stat(USERS_FILE).st_size == 0:
        with open(USERS_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["UserID", "Email", "Password"])
            writer.writerow(["U001", "test@example.com", "password123"])
            writer.writerow(["U002", "user1@example.com", "pass123"])
            writer.writerow(["U003", "user2@example.com", "pass456"])
            writer.writerow(["U004", "user3@example.com", "pass789"])
            writer.writerow(["U005", "user4@example.com", "passabc"])
            writer.writerow(["U006", "user5@example.com", "passdef"])

def authenticate_user():
    email = input("Enter your email: ")
    password = input("Enter your password: ")
    with open(USERS_FILE, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["Email"] == email and row["Password"] == password:
                print("Authentication successful!")
                return True
    print("Invalid credentials.")
    return False

def read_flights():
    with open(FLIGHT_FILE, mode='r') as file:
        return list(csv.DictReader(file))

def write_flights(flights):
    with open(FLIGHT_FILE, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=["FlightID", "Origin", "Destination", "Date", "TotalSeats", "BookedSeats", "Price", "Stops", "Duration", "BaggageAllowance"])
        writer.writeheader()
        writer.writerows(flights)

def book_seat(flight_id):
    if not authenticate_user():
        return
    flights = read_flights()
    for flight in flights:
        if flight["FlightID"] == flight_id:
            if int(flight["BookedSeats"]) < int(flight["TotalSeats"]):
                price = flight.get("Price", "N/A")
                print(f"Price: ${price}")
                confirm = input("Proceed with booking? (yes/no): ")
                if confirm.lower() == "yes":
                    flight["BookedSeats"] = str(int(flight["BookedSeats"]) + 1)
                    write_flights(flights)
                    print("Booking successful! Confirmation email sent.")
                else:
                    print("Booking cancelled.")
            else:
                print("Flight is fully booked.")
            return
    print("Flight not found.")

def search_flight():
    flights = read_flights()
    print("Filter options: 1. Price 2. Duration 3. Stops")
    choice = input("Choose filter: ")
    try:
        if choice == "1":
            flights.sort(key=lambda x: int(x.get("Price", 0)))
        elif choice == "2":
            flights.sort(key=lambda x: int(x.get("Duration", 0)))
        elif choice == "3":
            flights.sort(key=lambda x: int(x.get("Stops", 0)))
    except KeyError as e:
        print(f"Error: Missing key {e} in flight data.")
        return
    for flight in flights:
        print(flight)

def performance_report():
    flights = read_flights()
    total_flights = len(flights)
    total_booked = sum(int(flight["BookedSeats"]) for flight in flights)
    print(f"Total Flights: {total_flights}, Total Seats Booked: {total_booked}")


def main():
    initialize_csv()
    initialize_users()
    while True:
        print("\nFlight Booking System:")
        print("1. Search Flight")
        print("2. Book Seat")
        print("3. Performance Report")
        print("4. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            search_flight()
        elif choice == "2":
            fid = input("Enter Flight ID: ")
            book_seat(fid)
        elif choice == "3":
            performance_report()
        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
